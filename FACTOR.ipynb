Exploratory Factor Analysis
===============================
reduce a large number of variable

### Method
1. PCA using eigen decompositon
reduce variables into a smaller set of factors which "explain" the variance in the original variables.
2. factor analysis

### Option
NFACTORS: max number of factors should be extracted  
PREPLOT: factor plot  before rotation  
PLOT: factor plot after rotation  
ROTATE=VARIMAX: matains orthogonal(uncorrelated) nature of the factor and tries to get the original variables to load high on one of the factors and low on the rest. If your factors are orthogonal,and you then want to use them as independent variable in regression, no collinearity.  
SCREE: scree plot, help determine how many factors to keep in the analysis, look for "elbow"  
PRIOR SMC;: number of factor retained will be determined by the proportion of total variance explained

### Metric
Communalities: how much variance in the original variable is explained by the total of all the facors which are kept in the analysis
